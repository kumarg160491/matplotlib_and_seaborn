```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
x,y
```

```python
########Object oriented method#########
fig= plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig= plt.figure()
axes1=fig.add_axes([0.2,0.1,0.8,0.8])
axes2=fig.add_axes([0.3,0.4,0.3,0.4])
#l,b,w,h

axes1.plot(x,y,'r')
axes1.set_xlabel('X1 Label')
axes1.set_ylabel('Y1 Label')
axes1.set_title('Large Title')
axes2.plot(y,x,'b')
axes2.set_xlabel('X2 Label')
axes2.set_ylabel('Y2 Label')
axes2.set_title('Small Title')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,0.5,0.5])
ax.plot(x,y,lw=2,color='red',alpha=0.5,marker='*',markersize=5)
plt.tight_layout()
```

```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hex')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='kde')

#kde - kernel density estimation plot
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='reg')
```

```python
sns.pairplot(tips,hue="sex")
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill']) 
sns.rugplot(tips['total_bill'])

plt.tight_layout()
```

```python
sns.barplot(data=tips,x='sex',y='total_bill')
```

```python
tips.head()
```

```python
import numpy as np
sns.barplot(data=tips,x='smoker',y='total_bill',estimator=np.mean)
```

```python
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='smoker',split=True)
plt.tight_layout()
```

```python
flights=sns.load_dataset('flights')
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidths=1)
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```
